```python
SQL_SOURCE = 'https://ekh331.carto.com/api/v2/sql?q='

import urllib2
import urllib
import StringIO
import ast
import pandas as pd

def queryCartoDB(query, format='CSV', source=SQL_SOURCE):
    '''queries carto datasets from a given carto account
    Arguments: 
    query - string: a valid sql query string
    format - outlut format  OPTIONAL (default CSV)
    source - a valid sql api endpoint OPTIONAL (default carto fb55 account)
    Returns:
    the return of the sql query AS A STRING
    NOTES:
    designed for the carto API, tested only with CSV return format'''
    
    data = urllib.urlencode({'format': format, 'q': query})
    try:
        response = urllib2.urlopen(source, data)
    except urllib2.HTTPError, e:
        raise ValueError('\n'.join(ast.literal_eval(e.readline())['error']))
    except Exception:
        raise
    return response.read()
```

# Task 1 â€” Familiarize with SQL Clauses


â€¢Sort data by start_station_id, tripduration

- Only checking trips with duration <= 3 hours

â€¢ Only show the top/last 10 records (aka head and tail in SQL)

â€¢ List all unique start_station_id values

â€¢ Aggregation functions:
- Count the number of trips (aka wc -l in SQL)
- Find the average/min/max trip duratio

```python
#Sort data by start_station_id, tripduration
#- Only checking trips with duration <= 3 hours 

query_1 = '''
SELECT * FROM citibike
WHERE tripduration <= 10800
ORDER BY start_station_id, tripduration'''

pd.read_csv(StringIO.StringIO(queryCartoDB(query_1)), sep=',')
```

```python
#Only show the top/last 10 records (aka head and tail in SQL) 

query_2 = '''
SELECT * FROM citibike
WHERE tripduration <= 10800
ORDER BY start_station_id ASC, tripduration ASC
LIMIT 10'''


query_3 = '''SELECT * FROM citibike
WHERE tripduration <= 10800
ORDER BY start_station_id DESC, tripduration DESC
LIMIT 10'''

pd.read_csv(StringIO.StringIO(queryCartoDB(query_2)), sep=',')
```

```python
pd.read_csv(StringIO.StringIO(queryCartoDB(query_3)), sep=',')
```

```python
#List unique start station id values

query_4 = '''
SELECT DISTINCT(start_station_id)
AS distinct_station_id
FROM citibike
LIMIT 10'''

pd.read_csv(StringIO.StringIO(queryCartoDB(query_4)), sep=',')
```

```python
# Count number of trips and find min max avg trip duration

query_5 = '''
SELECT DISTINCT start_station_id,
COUNT(start_station_id) AS trip_count,
AVG(tripduration) AS trip_avg,
MAX(tripduration) AS trip_max,
MIN(tripduration) AS trip_min
FROM citibike
WHERE tripduration <=10800
GROUP BY start_station_id
LIMIT 10'''

pd.read_csv(StringIO.StringIO(queryCartoDB(query_5)), sep=',')
```

# Task 2 â€” Working with date/time

â€¢ Selecting trips started on Feb-02-2015 only

â€¢ Selecting trips started on the weekends
   
    - What are average trip duration during weekends?
â€¢ Can we do the same for weekday

```python
#Feb 02 2015 only

query_6 = '''
SELECT * FROM citibike
WHERE starttime >= '2015-02-02 00:00'
AND starttime < '2015-02-03 00:00'
LIMIT 10'''

#trips that start on weekends

query_7 = '''
SELECT * FROM citibike
WHERE EXTRACT(DOW FROM starttime) IN (0,6)
LIMIT 10'''

# average trip duration during weekends

query_8 = '''
SELECT start_station_id,
starttime,
usertype,
gender,
AVG(tripduration) AS trip_avg
FROM citibike
WHERE EXTRACT(DOW FROM starttime) IN (0,6)
GROUP BY start_station_id, starttime, usertype, gender
LIMIT 10'''

# average trip duration during weekdays

query_9 = '''
SELECT start_station_id,
starttime,
usertype,
gender,
AVG(tripduration) AS trip_avg
FROM citibike
WHERE EXTRACT(DOW FROM starttime) NOT IN (0,6)
GROUP BY start_station_id, starttime, usertype, gender
LIMIT 10'''

pd.read_csv(StringIO.StringIO(queryCartoDB(query_6)), sep=',')
```

# Task 3 â€” Working with Space

â€¢ Showing the list of start station locations
- Using GROUP BY

â€¢ Showing the number of trips started per station

â€¢ â€¦ but only for stations within 500m of Time Square!

- The coordinates of Time Square is (40.7577,-73.9857)


```python
# Station locations via groupby

query_10 ='''
SELECT CDB_TransformToWebmercator(CDB_LatLng(
start_station_latitude,
start_station_longitude)) 
AS THE_GEOM_WEBMERCATOR,
MIN(cartodb_id) as cartodb_id
FROM citibike

GROUP BY start_station_latitude, start_station_longitude
LIMIT 10
'''

# Number of trips started at each station

query_11 = '''
SELECT start_station_id,
CDB_TRANSFORMTOWEBMERCATOR(CDB_LatLng(
start_station_latitude,
start_station_longitude
))
AS THE_GEOM_WEBMERCATOR,
MIN(cartodb_id) AS carto_id,
COUNT(tripduration) AS trip_count
FROM citibike

GROUP BY start_station_id, start_station_latitude,
start_station_longitude
LIMIT 10'''


# Showing the number of trips started per station
# but only for stations within 500m of Time Square!
# The coordinates of Time Square is (40.7577,-73.9857)

query_12 = '''
SELECT start_station_id,
CDB_TRANSFORMTOWEBMERCATOR(CDB_LatLng(
start_station_latitude,
start_station_longitude
))
AS THE_GEOM_WEBMERCATOR,
MIN(cartodb_id) AS carto_id,
COUNT(tripduration) AS trip_count
FROM citibike
WHERE ST_DWITHIN(CDB_LatLng(
start_station_latitude,
start_station_longitude)::GEOGRAPHY,
CDB_LatLng(40.7577, -73.9857)::GEOGRAPHY, 500
)

GROUP BY start_station_id, start_station_latitude, start_station_longitude
LIMIT 10'''

pd.read_csv(StringIO.StringIO(queryCartoDB(query_10)), sep=',')
```

```python
pd.read_csv(StringIO.StringIO(queryCartoDB(query_12)), sep=',')
```

# Task 4 â€” Putting it all together

â€¢ Find the station that had the longest average trip duration during
weekends and within 500m of TimeSquare!


â€¢ Extra: create lines for trips started from stations within 500m of Times
Squares and lasted less than 2 hours. The number of trips per each
pair of stations are output as attributes of these lines. 

```python
query_13 = '''SELECT start_station_id,
starttime,
usertype,
gender,
AVG(tripduration) AS trip_avg
FROM citibike
WHERE EXTRACT(DOW FROM starttime) IN (0,6)
AND ST_DWITHIN(CDB_LatLng(
start_station_latitude,
start_station_longitude)::GEOGRAPHY,
CDB_LatLng(40.7577, -73.9857)::GEOGRAPHY, 500
)
GROUP BY start_station_id, starttime, usertype, gender
ORDER BY trip_avg DESC
LIMIT 10'''
```

```python
pd.read_csv(StringIO.StringIO(queryCartoDB(query_13)), sep=',')
```
